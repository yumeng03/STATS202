Exercise 1: 
(a) Flexible Statistical learning method will perform better than the inflexible method. As the sample size n is extremely large, the model will less likely to overfit, and the number of predictors p is small, an inflexible statistical learning model may lead to high bias situation and the model will not be that accurate.

(b) Flexible Statistical learning methoad will perform worse than inflexible method, because when p is large and n is small, the degree of complexity is large and it is more likely to overfit. Inflexible methods are efficient for large datasets with limited features.

(c) Flexible method performs better, flexible statistical method, also known as non-parametric methods, are able to adapt to the complexity of the data and can model a wide range of underlying data distributions, while inflexible methods make assumptions and are restricted to a range of underlying data distribution, and will always return estimate fhat that are close to linear.

(d) Flexible method performs worse. As there is a high variance, a flexible method might fit to the noise and increase the variance and lead to the problem of overfitting.
